# 4.8 Grouping Data & Aggregating Variables

```python
#Import pandas numpy and os
import pandas as pd
import numpy as np
import os
```

```python
#Import merged dataframe df_ords_prods_prior_merged as ords_prods_merge
ords_prods_merge = pd.read_pickle('/Users/ariedlin/Desktop/Career Foundry Assignments/8-24 Instacart Blanket Analysis Real/2 Data/2 Prepared Data/df_ords_prods_prior_merged')
```

# Grouping Data

```python
#Create a subset of data containging only the first 1,000,000 rows
df = ords_prods_merge[:1000000]
```

```python
#Check the head to make sure it worked
df.head()
```

```python
#Check the shape of the sample dataframe to ensure 1,000,000 rows
df.shape
```

# Grouping Data with pandas

```python
#Use the groupby function to group the 'product_name' column
df.groupby('product_name')
```

# Performing a Single Aggregation

```python
#Calculate the mean of the order_number and group by department_id
df.groupby('department_id').agg({'order_number': ['mean']})
```

```python
#Aggregate the data using the mean() function
df.groupby('department_id')['order_number'].mean()
```

# Performing Multiple Aggregations

```python
#Calculate the mean, min, and max of the order_number and group by department_id
df.groupby('department_id').agg({'order_number': ['mean', 'min', 'max']})
```

# Aggregating Data with transform()

```python
#Create new column 'max_order' to find max for 'order_number' and group by 'user_id'
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

# Deriving Columns with loc()

```python
#Assign 'loyalty' flags using loc()
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
#Check the value_counts of new 'loyalty_flag' column
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```
