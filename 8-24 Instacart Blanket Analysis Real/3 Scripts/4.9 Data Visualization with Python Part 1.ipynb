# Mixed Type Data

```python
##Check for mixed type data
for col in new_cust_data.columns:
    # Compare types in the column with the type of the first row
    weird = (new_cust_data[col].apply(type) != type(new_cust_data[col].iloc[0]))
    
    # If any row has a different type, print the column name
    if weird.any():
        print(col)
```
