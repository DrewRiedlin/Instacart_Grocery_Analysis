# Creating a function

```python
#Define a function to apply price levels to products in the dataframe

def price_label(row):

    if row['prices'] <= 5:
        return 'low range product'
    elif (row['prices'] > 5) and (row['prices'] <= 15):
        return 'mid range product'
    elif row['prices'] > 15:
        return 'high range product'
    else: return np.nan
```

```python
#Apply the fuction

df['price_range'] = df.apply(price_label, axis=1)
```

```python
#Check value counts
df['price_range'].value_counts(dropna = False)
```

# 2. Suppose your clients have changed their minds about the labels you created in your “busiest_day” column. Now, they want “Busiest day” to become “Busiest days” (plural). This label should correspond with the two busiest days of the week as opposed to the single busiest day. At the same time, they’d also like to know the two slowest days. Create a new column for this using a suitable method.

```python
#Use value_counts to find out to find frequency in the orders_dow column
df_ords_prods_prior_merged['orders_dow'].value_counts(dropna = False)
```

```python
# Create a for-loop assigning Saturday and Sunday as busiest days,
# Tuesday and Wednesday as least busy days, and all others as regularly busy
result = []

for value in df_ords_prods_prior_merged["orders_dow"]:
    if value in [0, 1]:  # Busiest days (Saturday, Sunday)
        result.append("Busiest days")
    elif value in [3, 4]:  # Least busy days (Tuesday, Wednesday)
        result.append("Least busy days")
    else:  # All other days
        result.append("Regularly busy days")
```

```python
#Add the column into the dataframe
df_ords_prods_prior_merged['busiest_days'] = result
```

```python
# Create a for-loop assigning 10, 11, 14, 15, 13, 12, 16, and 9 as 'Most orders',
# 17, 8, 18, 19, 20, 7, 21, and 22 as 'Average orders',
# and all others as 'Fewest orders'
result = []

for value in df_ords_prods_prior_merged["order_hour_of_day"]:
    if value in [10, 11, 14, 15, 13, 12, 16, 9]:  # Most orders
        result.append("Most orders")
    elif value in [17, 8, 18, 19, 20, 7, 21, 22]:  # Average orders
        result.append("Average orders")
    else:  # All other values represent Fewest orders
        result.append("Fewest orders")
```
